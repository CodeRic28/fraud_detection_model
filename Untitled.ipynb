```python
# Splitting the dataset
# X: Features
# Y: Targets

X = df.drop(columns=['isFraud','nameOrig','nameDest'],axis=1)
y = df['isFraud']
```

```python
# Feature Engineering
# diff: difference between new and old balance
X['balance_diff_orig'] = X['newbalanceOrig'] - X['oldbalanceOrg']
X['balance_diff_dest'] = X['newbalanceDest'] - X['oldbalanceDest']

X['time_of_day(hourly)'] = X['step'] % 24
X
```

```python
X['type'].unique()
```

```python
# Label Encoding
le = LabelEncoder()
X['type'] = le.fit_transform(X['type'])
```

```python
X
```

```python
# Scaling
columns_to_normalize = ['amount', 'oldbalanceOrg', 'newbalanceOrig','oldbalanceDest','newbalanceDest','balance_diff_orig','balance_diff_dest']

# Create an instance of Normalizer
normalizer = Normalizer(norm='l2')

# Transform the selected columns using L2 normalization
X[columns_to_normalize] = normalizer.transform(X[columns_to_normalize]).round(4)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=42,random_state=42)
```
